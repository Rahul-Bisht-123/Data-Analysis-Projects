## Data Cleaning

```python
df.isnull().sum()    #checking for missing values
```

-> 12 cells are empty in 'Amount' column \

```python
df.notnull().sum()
```

## Exploratory Data Analysis (E.D.A)

### Gender

```python
total_people = len(df['Gender'])
total_people
```

```python
df['Gender'].value_counts().sort_values(ascending=False).reset_index()
```

```python
# showing gender data using counterplot in seaborn

custom_palette={'M':'lightblue','F':'pink'}
sns.set_theme(style='darkgrid')

sns.set(rc={'figure.figsize':(10,2)})    #(width,height)

ax = sns.countplot(y=df['Gender'],
                   palette=custom_palette ,
                   hatch='/',
                   ec = 'black',
                   width=0.3)

for label in ax.containers:
    ax.bar_label(label)

plt.title('Number of Female vs Male Customers')
plt.show()
```

```python
# showing gender data in piechart

total_males=7832
total_females=3407
plt.figure(figsize=(4,3)) #width,height
plt.pie([total_males,total_females],labels=['Male','Female'],autopct='%.1f%%',explode=[0.1,0],colors=['lightblue','pink'],radius=6)
plt.title('Percentage of buyers based on Gender')
plt.axis('equal')
plt.show()
```

__*Gender Analysis*__
- __Female buyers = 69.7% (7832)__
- __Male buyers = 30.3% (3407)__
